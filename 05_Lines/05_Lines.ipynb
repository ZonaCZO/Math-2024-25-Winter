#### Task 1: Find the equation of the line through $A(1, 2)$ and $B(3, 4)$.

1. **Formula for a line through two points**:
   $$y - y_1 = \frac{y_2 - y_1}{x_2 - x_1}(x - x_1)$$
2. **Substitute $A(1, 2)$ and $B(3, 4)$**:
   $$y - 2 = \frac{4 - 2}{3 - 1}(x - 1) \quad \Rightarrow \quad y - 2 = \frac{2}{2}(x - 1) \quad \Rightarrow \quad y = x + 1$$
3. **Equation of the line**:
   $$y = x + 1$$

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Task 1: Line through points A(1, 2) and B(3, 4)
x1, y1 = 1, 2
x2, y2 = 3, 4

x_line1 = np.linspace(0, 4, 100)
y_line1 = x_line1 + 1  # Equation: y = x + 1

plt.figure(figsize=(8, 6))
plt.plot(x_line1, y_line1, label="Line through A(1,2) and B(3,4)")
plt.title("Task 1: Line through A(1,2) and B(3,4)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

#### Task 2: Find the equation of a line through $A(1, 2)$ and parallel to $y = 2x + 3$.

1. **Lines parallel to $y = 2x + 3$ have slope $m = 2$**.
2. **Using point-slope form**:
   $$y - y_1 = m(x - x_1)$$
   Substituting $A(1, 2)$ and $m = 2$:
   $$y - 2 = 2(x - 1) \quad \Rightarrow \quad y = 2x$$

---

```python
# Task 2: Line parallel to y = 2x + 3 through A(1, 2)
y_line2 = 2 * x_line1  # Equation: y = 2x

plt.figure(figsize=(8, 6))
plt.plot(x_line1, y_line2, label="Parallel to y=2x+3 through A(1,2)", linestyle="--")
plt.title("Task 2: Line parallel to y=2x+3 through A(1,2)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

#### Task 3: Find the equation of a line through $A(1, 2)$ and perpendicular to $y = 2x + 3$.

1. **The slope of a line perpendicular to $y = 2x + 3$ is the negative reciprocal of $m = 2$, which is $-\frac{1}{2}$.**
2. **Using point-slope form**:

#### Task 4: Find the intersection of $y = 2x + 3$ and $y = 3x + 2$, and calculate the angle between them.

#### Task 7: Distance from $A(1, 2)$ to $y = 2x + 3$.

1. **Formula for distance from point to line**:
   $$d = \frac{|ax_1 + by_1 + c|}{\sqrt{a^2 + b^2}}$$
   For $y = 2x + 3 \, \Rightarrow \, 2x - y + 3 = 0$, $a = 2, b = -1, c = 3$.
   Substituting $A(1, 2)$:
   $$d = \frac{|2 \cdot 1 - 1 \cdot 2 + 3|}{\sqrt{2^2 + (-1)^2}} = \frac{|3|}{\sqrt{5}} = \frac{3}{\sqrt{5}}$$

---


```python
# Task 7: Distance from point A(1, 2) to line y = 2x + 3

# Line: y = 2x + 3 rewritten as 2x - y + 3 = 0
A, B, C = 2, -1, 3  # Coefficients of the line
x_A, y_A = 1, 2     # Coordinates of point A

# Distance formula: |Ax + By + C| / sqrt(A^2 + B^2)
distance = abs(A * x_A + B * y_A + C) / np.sqrt(A**2 + B**2)
print(f"Task 7: Distance from A(1, 2) to line y = 2x + 3 is {distance}")
```
