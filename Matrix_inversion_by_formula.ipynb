# Problem 1: Finding the Inverse Matrix

Given matrix:

$$
M = \begin{bmatrix} 2 & 0 & 1 \\ 0 & 1 & 0 \\ 1 & 2 & 0 \end{bmatrix}
$$

We compute its inverse $$ M^{-1} $$ and verify the result by calculating $$ M \cdot M^{-1} $$.

The inverse of the matrix is:

$$
M^{-1} = \begin{bmatrix} \frac{4}{3} & 0 & -\frac{1}{3} \\ -\frac{2}{3} & 0 & \frac{1}{3} \\ 1 & 0 & 0 \end{bmatrix}
$$

Verification of the result:

$$
M \cdot M^{-1} = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}
$$

Thus, the inverse is correct.

```python
import sympy as sp

# Problem 1: Find the inverse of the matrix
matrix_1 = sp.Matrix([[2, 0, 1], [0, 1, 0], [1, 2, 0]])
inverse_matrix_1 = matrix_1.inv()  # Finding the inverse
verification = matrix_1 * inverse_matrix_1
verification  # Verifying the result
```

# Problem 2: Determinant and Rank of a Matrix

Given matrix:

$$
A = \begin{bmatrix} 4 & -3 & 7 \\ -1 & 6 & 3 \\ 2 & 9 & 1 \end{bmatrix}
$$

The determinant of matrix \( A \) is:

$$
\det(A) = -306
$$

Since the determinant is non-zero, the matrix has full rank. Therefore, the rank of matrix \( A \) is \( 3 \).

```python
import sympy as sp
matrix_A = sp.Matrix([[4, -3, 7], [-1, 6, 3], [2, 9, 1]])
det_A = matrix_A.det()  # Determinant
det_A
```

```python
rank_A = matrix_A.rank() if det_A != 0 else "Rank is less than full (singular matrix)"
rank_A
```
