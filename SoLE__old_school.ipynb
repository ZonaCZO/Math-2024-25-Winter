### Solving Systems of Equations:

#### System 1:
Given the equations:
$$ 3x - 2y = 5 \quad (1) $$
$$ 2x + 3y = 7 \quad (2) $$

**Step 1:** Express \( x \) from equation (1):

$$ x = \frac{5 + 2y}{3} $$

**Step 2:** Substitute \( x \) into equation (2):

$$ 2\left(\frac{5 + 2y}{3}\right) + 3y = 7 $$

Simplify the equation:

$$ \frac{2(5 + 2y)}{3} + 3y = 7 $$
$$ \frac{10 + 4y}{3} + 3y = 7 $$

Multiply by 3 to eliminate the denominator:

$$ 10 + 4y + 9y = 21 $$

**Step 3:** Solve for \( y \):

$$ 13y = 11 $$
$$ y = \frac{11}{13} $$

**Step 4:** Substitute \( y = \frac{11}{13} \) into the expression for \( x \):

$$ x = \frac{5 + 2 \times \frac{11}{13}}{3} = \frac{5 + \frac{22}{13}}{3} = \frac{\frac{65}{13} + \frac{22}{13}}{3} = \frac{\frac{87}{13}}{3} = \frac{87}{39} = \frac{29}{13} $$

**Solution:** $$ x = \frac{29}{13}, \quad y = \frac{11}{13} $$

#### System 2:
Given the equations:
$$ 2x - 3y = 10 \quad (1) $$
$$ 4x + 5y = 20 \quad (2) $$

**Step 1:** Multiply equation (1) by 2 to align with the coefficients of \( x \) in equation (2):

$$ 4x - 6y = 20 \quad (3) $$

**Step 2:** Subtract equation (2) from equation (3):

$$ (4x - 6y) - (4x + 5y) = 20 - 20 $$
$$ -11y = 0 $$

**Step 3:** Solve for \( y \):

$$ y = 0 $$

**Step 4:** Substitute \( y = 0 \) into equation (1):

$$ 2x - 3(0) = 10 $$
$$ 2x = 10 $$
$$ x = 5 $$

**Solution:** $$ x = 5, \quad y = 0 $$

---

#### System 3:
Given the equations:
$$ 2x - y + z = 3 \quad (1) $$
$$ x + 2y - z = 1 \quad (2) $$
$$ 3x - y + 2z = 11 \quad (3) $$

**Step 1:** Add equation (1) and equation (2) to eliminate \( z \):

$$
(2x - y + z) + (x + 2y - z) = 3 + 1
$$
This simplifies to:
$$
3x + y = 4 \quad (4)
$$
Now, we have a simpler equation (4) in terms of \( x \) and \( y \).

**Step 2:** Solve equation (4) for \( y \):
$$
y = 4 - 3x \quad (5)
$$

**Step 3:** Substitute \( y = 4 - 3x \) into equation (2) to eliminate \( y \):

$$
x + 2(4 - 3x) - z = 1
$$
Simplify:
$$
x + 8 - 6x - z = 1
$$
$$
-5x + 8 - z = 1
$$
$$
-5x - z = -7
$$
$$
z = 5x + 7 \quad (6)
$$

**Step 4:** Substitute \( y = 4 - 3x \) and \( z = 5x + 7 \) into equation (3):

$$
3x - (4 - 3x) + 2(5x + 7) = 11
$$
Simplify:
$$
3x - 4 + 3x + 10x + 14 = 11
$$
$$
16x + 10 = 11
$$
$$
16x = 1
$$
$$
x = \frac{1}{16}
$$

**Step 5:** Substitute \( x = \frac{1}{16} \) into equations (5) and (6) to find \( y \) and \( z \):

For \( y \):
$$
y = 4 - 3 \times \frac{1}{16} = 4 - \frac{3}{16} = \frac{64}{16} - \frac{3}{16} = \frac{61}{16}
$$

For \( z \):
$$
z = 5 \times \frac{1}{16} + 7 = \frac{5}{16} + \frac{112}{16} = \frac{117}{16}
$$

**Solution for System 3:**
$$
x = \frac{1}{16}, \quad y = \frac{61}{16}, \quad z = \frac{117}{16}
$$

#### System 4:
Given the equations:
$$ 2x - 3y + 4z + 2t = 2 \quad (1) $$
$$ 3x + 2y - 5z + 3t = 3 \quad (2) $$
$$ 4x - 3y + 2z - 5t = 4 \quad (3) $$
$$ 5x + 4y - 3z + 2t = 5 \quad (4) $$

**Step 1:** Add equation (1) and equation (2) to eliminate \( z \):

$$
(2x - 3y + 4z + 2t) + (3x + 2y - 5z + 3t) = 2 + 3
$$
Simplify:
$$
5x - y - z + 5t = 5 \quad (5)
$$

**Step 2:** Subtract equation (2) from equation (3) to eliminate \( z \):

$$
(4x - 3y + 2z - 5t) - (3x + 2y - 5z + 3t) = 4 - 3
$$
Simplify:
$$
x - 5y + 7z - 8t = 1 \quad (6)
$$

**Step 3:** Solve equation (5) for \( z \):

$$
z = 5x - y + 5t - 5 \quad (7)
$$

**Step 4:** Substitute \( z \) from equation (7) into equation (6):

$$
x - 5y + 7(5x - y + 5t - 5) - 8t = 1
$$
Simplify:
$$
x - 5y + 35x - 7y + 35t - 35 - 8t = 1
$$
$$
36x - 12y + 27t = 36 \quad (8)
$$
Divide by 3:
$$
12x - 4y + 9t = 12 \quad (9)
$$

**Step 5:** Now, substitute equation (7) into equation (1) and simplify to solve for \( x, y, t \).


```python
import sympy as sp

# System 1
x, y = sp.symbols('x y')
eq1_1 = sp.Eq(3*x - 2*y, 5)
eq1_2 = sp.Eq(2*x + 3*y, 7)
sol1 = sp.solve([eq1_1, eq1_2], (x, y))
print(f"Solution to System 1: {sol1}")

# System 2
eq2_1 = sp.Eq(2*x - 3*y, 10)
eq2_2 = sp.Eq(4*x + 5*y, 20)
sol2 = sp.solve([eq2_1, eq2_2], (x, y))
print(f"Solution to System 2: {sol2}")

# System 3
z = sp.symbols('z')
eq3_1 = sp.Eq(2*x - y + z, 3)
eq3_2 = sp.Eq(x + 2*y - z, 1)
eq3_3 = sp.Eq(3*x - y + 2*z, 11)
sol3 = sp.solve([eq3_1, eq3_2, eq3_3], (x, y, z))
print(f"Solution to System 3: {sol3}")

# System 4
z, t = sp.symbols('z t')
eq4_1 = sp.Eq(2*x - 3*y + 4*z + 2*t, 2)
eq4_2 = sp.Eq(3*x + 2*y - 5*z + 3*t, 3)
eq4_3 = sp.Eq(4*x - 3*y + 2*z - 5*t, 4)
eq4_4 = sp.Eq(5*x + 4*y - 3*z + 2*t, 5)
sol4 = sp.solve([eq4_1, eq4_2, eq4_3, eq4_4], (x, y, z, t))
print(f"Solution to System 4: {sol4}")
```
