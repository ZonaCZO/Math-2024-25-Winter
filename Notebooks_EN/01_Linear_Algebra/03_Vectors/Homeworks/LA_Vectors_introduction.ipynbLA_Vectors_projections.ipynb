#### Task 1: Divide the Line Segment

1. **Operation**: Divide the line segment connecting points $ A(-1, 2) $ and $ B(3, -2) $ in the ratio $ 1 : 3 $.
2. **Formula**: The formula for dividing a line segment in the ratio \( m : n \) is:
   $$
   P = \left( \frac{m \cdot x_2 + n \cdot x_1}{m + n}, \frac{m \cdot y_2 + n \cdot y_1}{m + n} \right)
   $$
   where $ P $ is the dividing point.
3. **Solution**: Substituting $ A(-1, 2) $, $ B(3, -2) $, $ m = 1 $, $ n = 3 $:

   $$
   P_x = \frac{1 \cdot 3 + 3 \cdot (-1)}{1 + 3} = \frac{3 - 3}{4} = 0, \quad
   P_y = \frac{1 \cdot (-2) + 3 \cdot 2}{1 + 3} = \frac{-2 + 6}{4} = 1
   $$

   Therefore, the dividing point is:
   
   $
   P(0, 1)
   $

---

```python
import matplotlib.pyplot as plt
import numpy as np

# Task 1: Divide the line segment
A = np.array([-1, 2])
B = np.array([3, -2])
ratio = (1, 3)
P = (ratio[0] * B + ratio[1] * A) / sum(ratio)

# Plot Task 1
plt.figure(figsize=(10, 6))
plt.plot([A[0], B[0]], [A[1], B[1]], 'b-o', label='Line Segment AB')
plt.scatter(*P, color='red', label=f'Dividing Point P{P}')
plt.text(A[0], A[1], 'A', fontsize=12, color='blue')
plt.text(B[0], B[1], 'B', fontsize=12, color='blue')
plt.text(P[0], P[1], f'P({P[0]:.1f}, {P[1]:.1f})', fontsize=12, color='red')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.legend()
plt.title("Task 1: Divide the Line Segment")
plt.show()

```

#### Task 2: Project Vector
$ \mathbf{a} $ onto $ OX $ and $ OY $

1. **Operation**: Project vector $ \mathbf{a} = (3, 4) $ onto the $ OX $-axis and the $ OY $-axis.
2. **Projection Formula**:
   - The projection onto \( OX \) is:
     $$
     \text{Proj}_{OX}(\mathbf{a}) = (a_x, 0)
     $$
   - The projection onto \( OY \) is:
     $$
     \text{Proj}_{OY}(\mathbf{a}) = (0, a_y)
     $$
3. **Solution**:
   - Onto \( OX \):
     $$
     \text{Proj}_{OX}(\mathbf{a}) = (3, 0)
     $$
   - Onto \( OY \):
     $$
     \text{Proj}_{OY}(\mathbf{a}) = (0, 4)
     $$


```python
# Task 2: Projections onto OX and OY
a = np.array([3, 4])
proj_OX = np.array([a[0], 0])
proj_OY = np.array([0, a[1]])

# Plot Task 2
plt.figure(figsize=(10, 6))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a')
plt.quiver(0, 0, proj_OX[0], proj_OX[1], angles='xy', scale_units='xy', scale=1, color='green', label='Proj onto OX')
plt.quiver(0, 0, proj_OY[0], proj_OY[1], angles='xy', scale_units='xy', scale=1, color='red', label='Proj onto OY')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.legend()
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.title("Task 2: Projections of Vector a onto OX and OY")
plt.show()
```

#### Task 3: Project Vector
$ \mathbf{a} $ onto $ \mathbf{b} $

1. **Operation**: Project vector $ \mathbf{a} = (2, 3) $ onto $ \mathbf{b} = (1, 1) $.
2. **Formula**: The projection of $ \mathbf{a} $ onto $ \mathbf{b} $ is:
   $$
   \text{Proj}_{\mathbf{b}}(\mathbf{a}) = \frac{\mathbf{a} \cdot \mathbf{b}}{\mathbf{b} \cdot \mathbf{b}} \mathbf{b}
   $$
3. **Solution**:
   - Dot products:
     $$
     \mathbf{a} \cdot \mathbf{b} = 2 \cdot 1 + 3 \cdot 1 = 5, \quad
     \mathbf{b} \cdot \mathbf{b} = 1^2 + 1^2 = 2
     $$
   - Substituting into the formula:
     $$
     \text{Proj}_{\mathbf{b}} $mathbf{a}$ = \frac{5}{2} \begin{bmatrix} 1 \\ 1 \end{bmatrix} is \begin{bmatrix} 2.5 \\ 2.5 \end{bmatrix}
     $$

---

```python
# Task 3: Project a onto b
a = np.array([2, 3])
b = np.array([1, 1])
proj_b_a = (np.dot(a, b) / np.dot(b, b)) * b

# Plot Task 3
plt.figure(figsize=(10, 6))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector b')
plt.quiver(0, 0, proj_b_a[0], proj_b_a[1], angles='xy', scale_units='xy', scale=1, color='red', label='Projection of a onto b')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.legend()
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.title("Task 3: Projection of a onto b")
plt.show()
```

#### Task 4: Project Vector
$ \mathbf{b} $ onto $ \mathbf{a} $

1. **Operation**: Project vector $ \mathbf{b} = (1, 1) $ onto $ \mathbf{a} = (2, 3) $.
2. **Formula**:
   $$
   \text{Proj}_{\mathbf{a}}(\mathbf{b}) = \frac{\mathbf{b} \cdot \mathbf{a}}{\mathbf{a} \cdot \mathbf{a}} \mathbf{a}
   $$
3. **Solution**:
   - Dot products:
     $$
     \mathbf{b} \cdot \mathbf{a} = 1 \cdot 2 + 1 \cdot 3 = 5, \quad
     \mathbf{a} \cdot \mathbf{a} = 2^2 + 3^2 = 13
     $$
   - Substituting into the formula:
     $$
     \text{Proj}_{\mathbf{a}}(\mathbf{b}) = \frac{5}{13} \begin{bmatrix} 2 \\ 3 \end{bmatrix} = \begin{bmatrix} \frac{10}{13} \\ \frac{15}{13} \end{bmatrix}
     $$

---

```python
# Task 4: Project b onto a
proj_a_b = (np.dot(b, a) / np.dot(a, a)) * a

# Plot Task 4
plt.figure(figsize=(10, 6))
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector b')
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector a')
plt.quiver(0, 0, proj_a_b[0], proj_a_b[1], angles='xy', scale_units='xy', scale=1, color='red', label='Projection of b onto a')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.legend()
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.title("Task 4: Projection of b onto a")
plt.show()
```
