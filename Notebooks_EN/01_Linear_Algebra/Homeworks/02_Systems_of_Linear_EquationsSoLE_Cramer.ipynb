# Solving Systems of Linear Equations using Cramer's Rule

## Problem 1

Solve the system of equations:
$$
\begin{cases}
2x_1 - 3x_2 = 7 \\
3x_1 + 5x_2 = 2
\end{cases}
$$

### Step 1: Write the coefficient matrix $A$ and the vector of constants $B$

The coefficient matrix $A$ is:
$$
A = \begin{pmatrix} 2 & -3 \\ 3 & 5 \end{pmatrix}
$$
The vector of constants $B$ is:
$$
B = \begin{pmatrix} 7 \\ 2 \end{pmatrix}
$$

### Step 2: Find the determinant of $A$

We calculate the determinant of $A$:
$$
\det(A) = 2 \times 5 - (-3) \times 3 = 10 + 9 = 19
$$
Since $\det(A) = 19 \neq 0$, the system has a unique solution.

### Step 3: Apply Cramer's Rule

Now we calculate $x_1$ and $x_2$ using Cramer's Rule:
$$
x_1 = \frac{\det(A_1)}{\det(A)}, \quad x_2 = \frac{\det(A_2)}{\det(A)}
$$

#### Find $A_1$:
Matrix $A_1$ is formed by replacing the first column of $A$ with $B$:
$$
A_1 = \begin{pmatrix} 7 & -3 \\ 2 & 5 \end{pmatrix}
$$
Now, calculate the determinant of $A_1$:
$$
\det(A_1) = 7 \times 5 - (-3) \times 2 = 35 + 6 = 41
$$

#### Find $A_2$:
Matrix $A_2$ is formed by replacing the second column of $A$ with $B$:
$$
A_2 = \begin{pmatrix} 2 & 7 \\ 3 & 2 \end{pmatrix}
$$
Now, calculate the determinant of $A_2$:
$$
\det(A_2) = 2 \times 2 - 7 \times 3 = 4 - 21 = -17
$$

### Step 4: Compute the solutions

Finally, we can find $x_1$ and $x_2$:
$$
x_1 = \frac{\det(A_1)}{\det(A)} = \frac{41}{19} \approx 2.16
$$
$$
x_2 = \frac{\det(A_2)}{\det(A)} = \frac{-17}{19} \approx -0.89
$$

Thus, the solution is:
$$
x_1 \approx 2.16, \quad x_2 \approx -0.89
$$

```python

import sympy as sp


x1, x2 = sp.symbols('x1 x2')
eq1 = sp.Eq(2*x1 - 3*x2, 7)
eq2 = sp.Eq(3*x1 + 5*x2, 2)

display(eq1, eq2)



```

```python
A1 = sp.Matrix([[2, -3],
               [3, 5]])
b1 = sp.Matrix([7, 2])

det_A1 = A1.det()


x1_sol = A1.copy()
x1_sol[:, 0] = b1
x2_sol = A1.copy()
x2_sol[:, 1] = b1

x1_value = x1_sol.det() / det_A1
x2_value = x2_sol.det() / det_A1

solution1 = (x1_value, x2_value)


solution1
```

## Problem 2

Solve the system of equations:
$$
\begin{cases}
2x + y - z = 1 \\
x - y + 2z = 4 \\
3x - 2z = -1
\end{cases}
$$

### Step 1: Write the coefficient matrix $A$ and the vector of constants $B$

The coefficient matrix $A$ is:
$$
A = \begin{pmatrix} 2 & 1 & -1 \\ 1 & -1 & 2 \\ 3 & 0 & -2 \end{pmatrix}
$$
The vector of constants $B$ is:
$$
B = \begin{pmatrix} 1 \\ 4 \\ -1 \end{pmatrix}
$$

### Step 2: Find the determinant of $A$

We calculate the determinant of $A$ using cofactor expansion:
$$
\det(A) = 2 \times \left( (-1) \times (-2) - 2 \times 0 \right) - 1 \times \left( 1 \times (-2) - 2 \times 3 \right) + (-1) \times \left( 1 \times 0 - (-1) \times 3 \right)
$$
$$
= 2 \times 2 - 1 \times (-2 - 6) + (-1) \times 3 = 4 - 1 \times (-8) - 3 = 4 + 8 - 3 = 9
$$

### Step 3: Apply Cramer's Rule

Now we calculate $x$, $y$, and $z$ using Cramer's Rule:
$$
x = \frac{\det(A_x)}{\det(A)}, \quad y = \frac{\det(A_y)}{\det(A)}, \quad z = \frac{\det(A_z)}{\det(A)}
$$

#### Find $A_x$:
Matrix $A_x$ is formed by replacing the first column of $A$ with $B$:
$$
A_x = \begin{pmatrix} 1 & 1 & -1 \\ 4 & -1 & 2 \\ -1 & 0 & -2 \end{pmatrix}
$$
Now, calculate the determinant of $A_x$ (similar to $A$).

```python
x, y, z = sp.symbols('x y z')

eq1 = sp.Eq(2*x + y - z, 1)
eq2 = sp.Eq(x - y + 2*z, 4)
eq3 = sp.Eq(3*x - 2*z, -1)

display(eq1, eq2, eq3)

A2 = sp.Matrix([[2, 1, -1],
               [1, -1, 2],
               [3, 0, -2]])
b2 = sp.Matrix([1, 4, -1])

det_A2 = A2.det()

if det_A2 != 0:
    x_sol = A2.copy()
    x_sol[:, 0] = b2
    y_sol = A2.copy()
    y_sol[:, 1] = b2
    z_sol = A2.copy()
    z_sol[:, 2] = b2

    x_value = x_sol.det() / det_A2
    y_value = y_sol.det() / det_A2
    z_value = z_sol.det() / det_A2

    solution2 = (x_value, y_value, z_value)
else:
    solution2 = "No unique solution (det(A) = 0)"

solution2
```

## Problem 3

Solve the system of equations:
$$
\begin{cases}
x + y + z - t = 2 \\
x - z + 2t = 6 \\
2x - 3y + t = 4 \\
3x + y + 3z - 4t = -2
\end{cases}
$$

### Step 1: Write the coefficient matrix $A$ and the vector of constants $B$

The coefficient matrix $A$ is:
$$
A = \begin{pmatrix} 1 & 1 & 1 & -1 \\ 1 & 0 & -1 & 2 \\ 2 & -3 & 0 & 1 \\ 3 & 1 & 3 & -4 \end{pmatrix}
$$
The vector of constants $B$ is:
$$
B = \begin{pmatrix} 2 \\ 6 \\ 4 \\ -2 \end{pmatrix}
$$

### Step 2: Find the determinant of $A$

We calculate the determinant of $A$ (details omitted here for brevity):
$$
\det(A) = 0
$$

Since $\det(A) = 0$, this system either has no solution or infinitely many solutions. Cramer's Rule cannot be applied because the matrix is singular.

---

```python

x, y, z, t = sp.symbols('x y z t')


eq1 = sp.Eq(x + y + z - t, 2)
eq2 = sp.Eq(x - z + 2*t, 6)
eq3 = sp.Eq(2*x - 3*y + t, 4)
eq4 = sp.Eq(3*x + y + 3*z - 4*t, -2)


display(eq1, eq2, eq3, eq4)

```

```python
A3 = sp.Matrix([[1, 1, 1, -1],
               [1, 0, -1, 2],
               [2, -3, 0, 1],
               [3, 1, 3, -4]])
b3 = sp.Matrix([2, 6, 4, -2])

det_A3 = A3.det()

if det_A3 != 0:
    x_sol = A3.copy()
    x_sol[:, 0] = b3
    y_sol = A3.copy()
    y_sol[:, 1] = b3
    z_sol = A3.copy()
    z_sol[:, 2] = b3
    t_sol = A3.copy()
    t_sol[:, 3] = b3

    x_value = x_sol.det() / det_A3
    y_value = y_sol.det() / det_A3
    z_value = z_sol.det() / det_A3
    t_value = t_sol.det() / det_A3

    solution3 = (x_value, y_value, z_value, t_value)
else:
    solution3 = "No unique solution (det(A) = 0)"

solution3
```

### Why can't the following system of equations be solved using Cramer's Rule?

$$
\begin{cases}
x_1 + 2x_2 + 3x_3 = 3 \\
4x_1 + 5x_2 + 6x_3 = 2 \\
7x_1 + 8x_2 + 9x_3 = 1
\end{cases}
$$

### Step 1: Write the coefficient matrix $A$

The coefficient matrix $A$ is:
$$
A = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{pmatrix}
$$

### Step 2: Find the determinant of $A$

We calculate the determinant of $A$:
$$
\det(A) = 1 \times \left( 5 \times 9 - 6 \times 8 \right) - 2 \times \left( 4 \times 9 - 6 \times 7 \right) + 3 \times \left( 4 \times 8 - 5 \times 7 \right)
$$
$$
= 1 \times (45 - 48) - 2 \times (36 - 42) + 3 \times (32 - 35)
$$
$$
= 1 \times (-3) - 2 \times (-6) + 3 \times (-3) = -3 + 12 - 9 = 0
$$

Since $\det(A) = 0$, the system cannot be solved using Cramer's Rule because the matrix is singular, meaning no unique solution exists.
