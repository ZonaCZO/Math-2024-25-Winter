### Problem 1:

Solve the system of linear equations using the inverse matrix method:

$$
\begin{cases}
x + 2y + 3z = 5 \\
2y + 3z = 4 \\
3z = 3
\end{cases}
$$

#### Step-by-step solution:

We can represent this system in matrix form as:
$$
A X = B
$$
Where:
$$
A = \begin{pmatrix}
1 & 2 & 3 \\
0 & 2 & 3 \\
0 & 0 & 3
\end{pmatrix}, \quad
X = \begin{pmatrix} x \\ y \\ z \end{pmatrix}, \quad
B = \begin{pmatrix} 5 \\ 4 \\ 3 \end{pmatrix}
$$

**Step 1:** Find the inverse of matrix \( A \):
$$
A^{-1} = \begin{pmatrix}
1 & 0 & 0 \\
0 & \frac{1}{2} & 0 \\
0 & 0 & \frac{1}{3}
\end{pmatrix}
$$

**Step 2:** Multiply \( A^{-1} \) with \( B \) to find \( X \):
$$
X = A^{-1} B
$$

Perform the multiplication:
$$
X = \begin{pmatrix}
1 & 0 & 0 \\
0 & \frac{1}{2} & 0 \\
0 & 0 & \frac{1}{3}
\end{pmatrix}
\begin{pmatrix} 5 \\ 4 \\ 3 \end{pmatrix}
= \begin{pmatrix}
5 \\
2 \\
1
\end{pmatrix}
$$

So, the solution is:
$$
x = 5, \quad y = 2, \quad z = 1
$$


```python
import sympy as sp

# Define variables
x, y, z = sp.symbols('x y z')

# Define the system of equations
eq1 = sp.Eq(x + 2*y + 3*z, 5)
eq2 = sp.Eq(2*y + 3*z, 4)
eq3 = sp.Eq(3*z, 3)

# Solve the system
solution = sp.solve([eq1, eq2, eq3], (x, y, z))
solution
```

### Problem 2:

Solve the system of linear equations using the inverse matrix method:

$$
\begin{cases}
x_1 + 2x_2 + 3x_3 = 41 \\
4x_1 + 5x_2 + 6x_3 = 93 \\
7x_1 + 8x_2 + 9x_3 = 145
\end{cases}
$$

#### Step-by-step solution:

We represent this system in matrix form:
$$
A X = B
$$
Where:
$$
A = \begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{pmatrix}, \quad
X = \begin{pmatrix} x_1 \\ x_2 \\ x_3 \end{pmatrix}, \quad
B = \begin{pmatrix} 41 \\ 93 \\ 145 \end{pmatrix}
$$

**Step 1:** Find the inverse of matrix \( A \).

Since the determinant of \( A \) is zero (it's a singular matrix), the system cannot be solved using the inverse matrix method.


```python
import sympy as sp

# Define variables
x1, x2, x3 = sp.symbols('x1 x2 x3')

# Define the system of equations
eq1 = sp.Eq(x1 + 2*x2 + 3*x3, 41)
eq2 = sp.Eq(4*x1 + 5*x2 + 6*x3, 93)
eq3 = sp.Eq(7*x1 + 8*x2 + 9*x3, 145)

# Attempt to solve the system
solution = sp.solve([eq1, eq2, eq3], (x1, x2, x3))
solution
```
