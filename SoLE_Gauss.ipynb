## Problem 1

We are given the system of equations:

$$
\begin{cases}
x + 2y - 2z = 4 \\
2x + y + z = 0 \\
3x + 2y + z = 1
\end{cases}
$$

### Step 1: Form the augmented matrix

The augmented matrix is:

$$
\begin{pmatrix}
1 & 2 & -2 & 4 \\
2 & 1 & 1 & 0 \\
3 & 2 & 1 & 1
\end{pmatrix}
$$

### Step 2: Apply Gaussian elimination to reduce the matrix to row echelon form

We perform elementary row operations to reduce the matrix.

1. Subtract $2 \times$ (row 1) from (row 2).
2. Subtract $3 \times$ (row 1) from (row 3).

$$
\begin{pmatrix}
1 & 2 & -2 & 4 \\
0 & -3 & 5 & -8 \\
0 & -4 & 7 & -11
\end{pmatrix}
$$

3. Multiply row 2 by $-\frac{1}{3}$:

$$
\begin{pmatrix}
1 & 2 & -2 & 4 \\
0 & 1 & -\frac{5}{3} & \frac{8}{3} \\
0 & -4 & 7 & -11
\end{pmatrix}
$$

4. Add $4 \times$ (row 2) to (row 3):

$$
\begin{pmatrix}
1 & 2 & -2 & 4 \\
0 & 1 & -\frac{5}{3} & \frac{8}{3} \\
0 & 0 & \frac{1}{3} & \frac{1}{3}
\end{pmatrix}
$$

### Step 3: Back substitution

Now we can use back substitution to find the solution.

From row 3:

$$
\frac{1}{3}z = \frac{1}{3} \implies z = 1
$$

From row 2:

$$
y - \frac{5}{3}z = \frac{8}{3} \implies y - \frac{5}{3} \times 1 = \frac{8}{3} \implies y = \frac{13}{3}
$$

From row 1:

$$
x + 2y - 2z = 4 \implies x + 2 \times \frac{13}{3} - 2 \times 1 = 4 \implies x = \frac{-6}{3} = -2
$$

Thus, the solution is:

$$
x = -2, \quad y = \frac{13}{3}, \quad z = 1
$$

## Problem 2

We are given the system of equations:

$$
\begin{cases}
x + y + z - t = 2 \\
2x + y + z = 3 \\
-x + z - t = 0 \\
3x + 2y - z + 2t = -1
\end{cases}
$$

### Step 1: Form the augmented matrix

We can represent this system as the augmented matrix:

$$
\begin{pmatrix}
1 & 1 & 1 & -1 & 2 \\
2 & 1 & 1 & 0 & 3 \\
-1 & 0 & 1 & -1 & 0 \\
3 & 2 & -1 & 2 & -1
\end{pmatrix}
$$

### Step 2: Apply Gaussian elimination to reduce the matrix to row echelon form

1. First, subtract $2 \times$ (row 1) from (row 2), and subtract $-1 \times$ (row 1) from (row 3). We get:

$$
\begin{pmatrix}
1 & 1 & 1 & -1 & 2 \\
0 & -1 & -1 & 2 & -1 \\
0 & 1 & 2 & -2 & 2 \\
3 & 2 & -1 & 2 & -1
\end{pmatrix}
$$

2. Subtract $3 \times$ (row 1) from (row 4):

$$
\begin{pmatrix}
1 & 1 & 1 & -1 & 2 \\
0 & -1 & -1 & 2 & -1 \\
0 & 1 & 2 & -2 & 2 \\
0 & -1 & -4 & 5 & -7
\end{pmatrix}
$$

3. Multiply row 2 by $-1$:

$$
\begin{pmatrix}
1 & 1 & 1 & -1 & 2 \\
0 & 1 & 1 & -2 & 1 \\
0 & 1 & 2 & -2 & 2 \\
0 & -1 & -4 & 5 & -7
\end{pmatrix}
$$

4. Subtract row 2 from row 3 and add row 2 to row 4:

$$
\begin{pmatrix}
1 & 1 & 1 & -1 & 2 \\
0 & 1 & 1 & -2 & 1 \\
0 & 0 & 1 & 0 & 1 \\
0 & 0 & -3 & 3 & -6
\end{pmatrix}
$$

5. Divide row 4 by $-3$:

$$
\begin{pmatrix}
1 & 1 & 1 & -1 & 2 \\
0 & 1 & 1 & -2 & 1 \\
0 & 0 & 1 & 0 & 1 \\
0 & 0 & 1 & -1 & 2
\end{pmatrix}
$$

6. Subtract row 3 from row 4:

$$
\begin{pmatrix}
1 & 1 & 1 & -1 & 2 \\
0 & 1 & 1 & -2 & 1 \\
0 & 0 & 1 & 0 & 1 \\
0 & 0 & 0 & -1 & 1
\end{pmatrix}
$$

### Step 3: Back substitution

1. From row 4:

$$
-t = 1 \implies t = -1
$$

2. From row 3:

$$
z = 1
$$

3. From row 2:

$$
y + z - 2t = 1 \implies y + 1 - 2(-1) = 1 \implies y = -2
$$

4. From row 1:

$$
x + y + z - t = 2 \implies x + (-2) + 1 - (-1) = 2 \implies x = 2
$$

Thus, the solution is:

$$
x = 2, \quad y = -2, \quad z = 1, \quad t = -1
$$


```python
import sympy as sp

# Define the variables
x, y, z, t = sp.symbols('x y z t')

# Define the augmented matrix
augmented_matrix = sp.Matrix([
    [1, 1, 1, -1, 2],
    [2, 1, 1, 0, 3],
    [-1, 0, 1, -1, 0],
    [3, 2, -1, 2, -1]
])

# Perform Gaussian elimination
rref_matrix, pivots = augmented_matrix.rref()
rref_matrix
```

## Problem 3

We are given the system of equations:

$$
\begin{cases}
x + y - z - t = 0 \\
2x + 3y - 2z + t = 4 \\
3x + 5z = 0 \\
-x + y - 3z + 2t = 3
\end{cases}
$$

### Step 1: Form the augmented matrix

The augmented matrix is:

$$
\begin{pmatrix}
1 & 1 & -1 & -1 & 0 \\
2 & 3 & -2 & 1 & 4 \\
3 & 0 & 5 & 0 & 0 \\
-1 & 1 & -3 & 2 & 3
\end{pmatrix}
$$

### Step 2: Apply Gaussian elimination to reduce the matrix to row echelon form

1. Subtract $2 \times$ (row 1) from (row 2), subtract $3 \times$ (row 1) from (row 3), and add (row 1) to (row 4):

$$
\begin{pmatrix}
1 & 1 & -1 & -1 & 0 \\
0 & 1 & 0 & 3 & 4 \\
0 & -3 & 8 & 3 & 0 \\
0 & 2 & -4 & 1 & 3
\end{pmatrix}
$$

2. Add $3 \times$ (row 2) to row 3 and subtract $2 \times$ (row 2) from row 4:

$$
\begin{pmatrix}
1 & 1 & -1 & -1 & 0 \\
0 & 1 & 0 & 3 & 4 \\
0 & 0 & 8 & 12 & 12 \\
0 & 0 & -4 & -5 & -5
\end{pmatrix}
$$

3. Multiply row 3 by $\frac{1}{8}$ and row 4 by $-\frac{1}{4}$:

$$
\begin{pmatrix}
1 & 1 & -1 & -1 & 0 \\
0 & 1 & 0 & 3 & 4 \\
0 & 0 & 1 & 1.5 & 1.5 \\
0 & 0 & 1 & 1.25 & 1.25
\end{pmatrix}
$$

4. Subtract row 3 from row 4:

$$
\begin{pmatrix}
1 & 1 & -1 & -1 & 0 \\
0 & 1 & 0 & 3 & 4 \\
0 & 0 & 1 & 1.5 & 1.5 \\
0 & 0 & 0 & -0.25 & -0.25
\end{pmatrix}
$$

### Step 3: Back substitution

1. From row 4:

$$
-0.25t = -0.25 \implies t = 1
$$

2. From row 3:

$$
z + 1.5t = 1.5 \implies z + 1.5 \times 1 = 1.5 \implies z = 0
$$

3. From row 2:

$$
y + 3t = 4 \implies y + 3 \times 1 = 4 \implies y = 1
$$

4. From row 1:

$$
x + y - z - t = 0 \implies x + 1 - 0 - 1 = 0 \implies x = 0
$$

Thus, the solution is:

$$
x = 0, \quad y = 1, \quad z = 0, \quad t = 1
$$


```python
# Define the variables
x, y, z, t = sp.symbols('x y z t')

# Define the augmented matrix
augmented_matrix3 = sp.Matrix([
    [1, 1, -1, -1, 0],
    [2, 3, -2, 1, 4],
    [3, 0, 5, 0, 0],
    [-1, 1, -3, 2, 3]
])

# Perform Gaussian elimination
rref_matrix3, pivots3 = augmented_matrix3.rref()
rref_matrix3
```
