1. Calculate the sum of matrices $A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} \ and \ B = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix}$.

$
A + B = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} + \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix} = \begin{pmatrix} 1 + 5 & 2 + 6 \\ 3 + 7 & 4 + 8 \end{pmatrix} = \begin{pmatrix} 6 & 8 \\ 10 & 12 \end{pmatrix}
$

2.
Calculate the product of matrix and scalar (k)

$$
A = \begin{pmatrix} 4 & 2 \\ 12 & 4 \end{pmatrix}, \quad k = \frac{1}{4}
$$

Multiply each element of the matrix by a scalar \( k \):

$$
kA = \frac{1}{4} \begin{pmatrix} 4 & 2 \\ 12 & 4 \end{pmatrix} = \begin{pmatrix} \frac{1}{4} \times 4 & \frac{1}{4} \times 2 \\ \frac{1}{4} \times 12 & \frac{1}{4} \times 4 \end{pmatrix}
$$

Result:

$$
kA = \begin{pmatrix} 1 & 0.5 \\ 3 & 1 \end{pmatrix}
$$

```python
# 2
import sympy as sp
A = sp.Matrix([[4, 2], [12, 4]])
k = 1/4

# Calculate the product of matrix A and scalar k
product = k * A
product
```
