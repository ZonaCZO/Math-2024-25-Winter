1. Calculate the determinants of the following 2x2 matrices:

$$
\mathbf{A} =
\begin{pmatrix}
2 & 3 \\
1 & 4
\end{pmatrix}
, \qquad
\mathbf{B} =
\begin{pmatrix}
5 & 6 \\
7 & 8
\end{pmatrix}
, \qquad
\mathbf{C} =
\begin{pmatrix}
-1 & 2 \\
3 & 0
\end{pmatrix}
$$

For matrix \( A \):

$$
\det(A) = 2 \times 4 - 3 \times 1 = 8 - 3 = 5
$$

For matrix \( B \):

$$
\det(B) = 5 \times 8 - 6 \times 7 = 40 - 42 = -2
$$

For matrix \( C \):

$$
\det(C) = (-1) \times 0 - 2 \times 3 = 0 - 6 = -6
$$

Thus, we have:
$$
\det(A) = 5, \quad \det(B) = -2, \quad \det(C) = -6
$$


```python
import sympy as sp

# Define matrices A, B, C
A_2x2 = sp.Matrix([[2, 3], [1, 4]])
B_2x2 = sp.Matrix([[5, 6], [7, 8]])
C_2x2 = sp.Matrix([[-1, 2], [3, 0]])

# Calculate determinants
det_A_2x2 = A_2x2.det()
det_B_2x2 = B_2x2.det()
det_C_2x2 = C_2x2.det()

det_A_2x2, det_B_2x2, det_C_2x2
```

2. Calculate the determinants of the following 3x3 matrices:

$$
\mathbf{D} =
\begin{pmatrix}
1 & 0 & 2 \\
-1 & 3 & 1 \\
2 & 4 & -2
\end{pmatrix}
, \qquad
\mathbf{E} =
\begin{pmatrix}
3 & 1 & -1 \\
0 & 2 & 4 \\
5 & 3 & 2
\end{pmatrix}
, \qquad
\mathbf{F} =
\begin{pmatrix}
2 & -3 & 1 \\
1 & 4 & -2 \\
1 & 5 & 3
\end{pmatrix}
$$

For matrix \( D \):

$$
\det(D) = 1 \cdot \begin{vmatrix} 3 & 1 \\ 4 & -2 \end{vmatrix} - 0 \cdot \begin{vmatrix} -1 & 1 \\ 2 & -2 \end{vmatrix} + 2 \cdot \begin{vmatrix} -1 & 3 \\ 2 & 4 \end{vmatrix}
$$

$$
\det(D) = 1(3 \times -2 - 1 \times 4) + 2((-1 \times 4 - 3 \times 2)) = 1(-6 - 4) + 2(-4 - 6) = -10 - 20 = -30
$$

For matrix \( E \):

$$
\det(E) = 3 \cdot \begin{vmatrix} 2 & 4 \\ 3 & 2 \end{vmatrix} - 1 \cdot \begin{vmatrix} 0 & 4 \\ 5 & 2 \end{vmatrix} + (-1) \cdot \begin{vmatrix} 0 & 2 \\ 5 & 3 \end{vmatrix}
$$

$$
\det(E) = 3(2 \times 2 - 4 \times 3) - 1(0 \times 2 - 4 \times 5) + (-1)(0 \times 3 - 2 \times 5) = 3(4 - 12) - 1(0 - 20) + (-1)(0 - 10) = -24 + 20 - 10 = -14
$$

For matrix \( F \):

$$
\det(F) = 2 \cdot \begin{vmatrix} 4 & -2 \\ 5 & 3 \end{vmatrix} - (-3) \cdot \begin{vmatrix} 1 & -2 \\ 1 & 3 \end{vmatrix} + 1 \cdot \begin{vmatrix} 1 & 4 \\ 1 & 5 \end{vmatrix}
$$

$$
\det(F) = 2(4 \times 3 - (-2) \times 5) - (-3)((1 \times 3 - (-2) \times 1)) + 1(1 \times 5 - 4 \times 1) = 2(12 + 10) + 3(3 + 2) + 1(5 - 4) = 44 + 15 + 1 = 60
$$

Thus, we have:
$$
\det(D) = -30, \quad \det(E) = -14, \quad \det(F) = 60
$$


```python
# Define matrices D, E, F
D_3x3 = sp.Matrix([[1, 0, 2], [-1, 3, 1], [2, 4, -2]])
E_3x3 = sp.Matrix([[3, 1, -1], [0, 2, 4], [5, 3, 2]])
F_3x3 = sp.Matrix([[2, -3, 1], [1, 4, -2], [1, 5, 3]])

# Calculate determinants
det_D_3x3 = D_3x3.det()
det_E_3x3 = E_3x3.det()
det_F_3x3 = F_3x3.det()

det_D_3x3, det_E_3x3, det_F_3x3
```
