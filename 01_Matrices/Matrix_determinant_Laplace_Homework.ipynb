Calculate the determinant of matrix \( A \):

$$
A = \begin{pmatrix} 2 & 3 & 1 \\ 1 & 4 & 0 \\ 3 & 2 & 1 \end{pmatrix}
$$

We calculate the determinant:

$$
\det(A) = 2 \begin{vmatrix} 4 & 0 \\ 2 & 1 \end{vmatrix} - 3 \begin{vmatrix} 1 & 0 \\ 3 & 1 \end{vmatrix} + 1 \begin{vmatrix} 1 & 4 \\ 3 & 2 \end{vmatrix}
$$

$$
\det(A) = 2(4 \cdot 1 - 0 \cdot 2) - 3(1 \cdot 1 - 0 \cdot 3) + 1(1 \cdot 2 - 4 \cdot 3)
$$

$$
\det(A) = 2(4) - 3(1) + 1(-10) = 8 - 3 - 10 = -5
$$




```python
import sympy as sp

# Define matrix A
A = sp.Matrix([[2, 3, 1], [1, 4, 0], [3, 2, 1]])

# Calculate determinant of A
det_A = A.det()
det_A
```

Calculate the determinant of matrix \( C \):

$$
C = \begin{pmatrix} 2 & 3 & 1 & 4 \\ 1 & 0 & 0 & 6 \\ 3 & 2 & 1 & 5 \\ 2 & 1 & 4 & 0 \end{pmatrix}
$$

Using cofactor expansion along the first row:

$$
\det(C) = 2 \begin{vmatrix} 0 & 0 & 6 \\ 2 & 1 & 5 \\ 1 & 4 & 0 \end{vmatrix} - 3 \begin{vmatrix} 1 & 0 & 6 \\ 3 & 1 & 5 \\ 2 & 4 & 0 \end{vmatrix} + 1 \begin{vmatrix} 1 & 0 & 6 \\ 3 & 2 & 5 \\ 2 & 1 & 0 \end{vmatrix} - 4 \begin{vmatrix} 1 & 0 & 0 \\ 3 & 2 & 1 \\ 2 & 1 & 4 \end{vmatrix}
$$

Solving this gives:

$$
\det(C) = -75
$$




```python
# Define matrix C
C = sp.Matrix([[2, 3, 1, 4], [1, 0, 0, 6], [3, 2, 1, 5], [2, 1, 4, 0]])

# Calculate determinant of C
det_C = C.det()
det_C
```

Calculate the determinant of matrix \( D \):

$$
D = \begin{pmatrix} 2 & 3 & 1 & 4 & 5 \\ 1 & 4 & 0 & 0 & 7 \\ 3 & 0 & 0 & 0 & 0 \\ 2 & 1 & 4 & 3 & 2 \\ 1 & 2 & 3 & 4 & 5 \end{pmatrix}
$$

Expanding by the third row (since it contains many zeros):

$$
\det(D) = 3 \cdot \det\begin{pmatrix} 3 & 1 & 4 & 5 \\ 4 & 0 & 0 & 7 \\ 1 & 4 & 3 & 2 \\ 2 & 3 & 4 & 5 \end{pmatrix}
$$

Further calculations yield:

$$
\det(D) = 231
$$




```python
# Define matrix D
D = sp.Matrix([[2, 3, 1, 4, 5], [1, 4, 0, 0, 7], [3, 0, 0, 0, 0], [2, 1, 4, 3, 2], [1, 2, 3, 4, 5]])

# Calculate determinant of D
det_D = D.det()
det_D
```
